```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./prob-full-matrix", index_col=0, names=np.arange(0, 20, 1, dtype=int))
df
```

```python
PROB_MATRIX = df.to_numpy()

for i in range(PROB_MATRIX.shape[1]):
    mean = np.sum(PROB_MATRIX[:, i]) / PROB_MATRIX.shape[1]

    PROB_MATRIX[:, i] = PROB_MATRIX[:, i] - mean



COV_MATRIX = PROB_MATRIX.T @ PROB_MATRIX

U, S, V = np.linalg.svd(COV_MATRIX)

print(U)
```

```python
print(S)
```

```python
first_pc = U[:, 1]

first_pc
```

```python
relative_probs = (first_pc.reshape(1, 20) @ PROB_MATRIX.T)

relative_probs
```

```python
dict = {"Class":df.index.to_numpy(), "Relative Probs": relative_probs.flatten()}

r_prob_df = pd.DataFrame(dict)

r_prob_df
```

```python
def sig(z):
    return 1 / (1 + np.exp(-z))


dict = {"Class":df.index.to_numpy(), "Probs": sig(relative_probs.flatten())}

r_prob_df = pd.DataFrame(dict)

r_prob_df
```
